```python
Car_Popularity = data.value_counts(["Make","Model"],ascending=True)
Car_Popularity
```

**It shows the Car Model Popularity Graph for the above models**

```python
Car_Popularity.tail(10).plot(kind='barh',title = 'Car Model Popularity Graph',figsize=(20,10), ylabel = "TOP 10 POPULAR MODELS", xlabel = "Total number of cars", fontsize = 15)
```

**It shows the Company Popularity Graph for the above models**

```python
Car_Popularity.tail(10).plot(kind='barh', title = 'Company Popularity Graph',figsize=(20,10),fontsize = 15, ylabel = "TOP 10 Popular Companies", xlabel = "Total number of cars")
```

**Creating a variable to identify the cars count per city**

```python
import pandas as pd
import numpy as np
CarsPerCounty = data.value_counts(["County"],ascending=True)
CarsPerCounty.tail(10).plot(kind='barh', title = 'Cars Per County Graph',figsize=(15,10),fontsize = 15, ylabel = "County", xlabel = "Total number of cars")
```

**Denotes the car per city**

```python
CarsPerCity = data.value_counts(["City"],ascending = True)
CarsPerCity.tail(10).plot(kind='barh', title = 'Cars Per City Graph',figsize=(15,10),fontsize = 15, ylabel = "City", xlabel = "Total number of cars")
```

** Creates an horizontal bar graph to indicate CarsPerCityandCounty**

```python
CarsPerCityandCounty = data.value_counts(["City","County"],ascending = True)
```

```python
CarsPerCityandCounty.tail(10).plot(kind='barh', title = 'Cars Per City and County Graph',figsize=(15,10),fontsize = 15, ylabel = "City and County", xlabel = "Total number of cars")
```
