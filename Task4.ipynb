LABEL ENCODING
Label Encoder class can only be used to encode categorical variables, and the "price(in lakhs)" column is a continuous variable. By encoding the values in the "price(in lakhs)" column, we can use the LabelEncoder class to transform the data into a format that is more suitable for machine learning algorithms.

In simple terms, this code converts the categorical values in the 'registration_year' column into numerical labels. Each unique registration year is assigned a unique numerical label, allowing the model to work with this feature effectively.

```python
import sklearn.preprocessing
label_encoder = sklearn.preprocessing.LabelEncoder()
car['registration_year']=label_encoder.fit_transform(car['registration_year'])
car['insurance_validity'] = label_encoder.fit_transform(car['insurance_validity'])
car['fuel_type'] = label_encoder.fit_transform(car['fuel_type'])
car['seats'] = label_encoder.fit_transform(car['seats'])
car['kms_driven'] = label_encoder.fit_transform(car['kms_driven'])
car['ownsership'] = label_encoder.fit_transform(car['ownsership'])
car['transmission'] = label_encoder.fit_transform(car['transmission'])
car['manufacturing_year'] = label_encoder.fit_transform(car['manufacturing_year'])
car['mileage(kmpl)'] = label_encoder.fit_transform(car['mileage(kmpl)'])
car['engine(cc)'] = label_encoder.fit_transform(car['engine(cc)'])
car['max_power(bhp)'] = label_encoder.fit_transform(car['max_power(bhp)'])
car['torque(Nm)'] = label_encoder.fit_transform(car['torque(Nm)'])
car['price(in lakhs)'] = label_encoder.fit_transform(car['price(in lakhs)'])
#car['company'] = label_encoder.fit_transform(car['comapny'])
```

The code imports the "Lasso" class from scikit-learn's linear_model module. It creates an instance of the Lasso model with a specified regularization parameter 'alpha', which controls the strength of the penalty for large coefficients. In this case, alpha is set to 0.2. The model is trained or "fit" to the training data (x_train, y_train). This involves finding the coefficients that minimize the Lasso regression objective function. The trained model is then used to predict the response variable for new data points ('x_test'), and the predicted values are stored in the variable "y_pred".

```python
from sklearn.linear_model import Lasso
model = Lasso(alpha=0.2)
model.fit(x_train,y_train)
y_pred = model.predict(x_test)
y_pred
```

This code calculates and prints the R^2 score for your Lasso regression model based on the provided test data (x_test, y_test). The R^2 score will give you an idea of how well your model is performing on the test set. If the R^2 score is closer to 1, it indicates a better fit of the model to the test data.

```python
model.score(x_test,y_test)
```
